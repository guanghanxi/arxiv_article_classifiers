```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
mlp_data = np.load("MLP_results.npz")
gru_data = np.load("GRU_results.npz")
```

```python
mlp_data['arr_0']
```

```python
plt.figure(figsize=(10, 8))
plt.plot(mlp_data['arr_0'][1], mlp_data['arr_0'][0], 'g-', linewidth = 2, label = 'MLP Train Loss')
plt.plot(gru_data['arr_0'][1], gru_data['arr_0'][0], 'r-', linewidth = 2, label = 'GRU Train Loss')
plt.plot(mlp_data['arr_1'][2], mlp_data['arr_1'][1], 'k*-', markevery = 5, markersize = 10,
         markerfacecolor = 'none', linewidth = 3, label = 'MLP Evaluation Loss')
plt.plot(gru_data['arr_1'][2], gru_data['arr_1'][1], 'bo-', markevery = 5, markersize = 10,
         markerfacecolor = 'none', linewidth = 3, label = 'GRU Evaluation Loss')
plt.legend(fontsize =16, loc = 'upper right');
plt.xlabel("Train Step", fontsize =18)
plt.ylabel("Loss", fontsize =18)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.savefig('loss.eps')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
plt.plot(mlp_data['arr_1'][2], mlp_data['arr_1'][1], 'g-', linewidth = 3, label = 'MLP Evaluation Loss')
plt.plot(gru_data['arr_1'][2], gru_data['arr_1'][1], 'r-', linewidth = 3, label = 'GRU Evaluation Loss')
plt.legend(fontsize =16, loc = 'upper right');
plt.xlabel("Train Step", fontsize =18)
plt.ylabel("Evaluation Loss", fontsize =18)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.savefig('eval_loss.eps')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
plt.plot(mlp_data['arr_1'][2], mlp_data['arr_1'][0], 'g-', linewidth = 3, label = 'MLP Evaluation Accuracy')
plt.plot(gru_data['arr_1'][2], gru_data['arr_1'][0], 'r-', linewidth = 3, label = 'GRU Evaluation Accuracy')
plt.legend(fontsize =16, loc = 'lower right');
plt.xlabel("Train Step", fontsize =18)
plt.ylabel("Evaluation Accuracy", fontsize =18)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.savefig('eval_acc.eps')
plt.show()
```
